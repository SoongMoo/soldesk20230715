```python
rainfall = pd.read_csv("https://raw.githubusercontent.com/SoongMoo/soldesk20230715/main/%EC%B0%B8%EA%B3%A0%EC%9E%90%EB%A3%8C/NumPy/data/Seattle2014.csv")
rainfall['PRCP']
inches = np.array(rainfall['PRCP'] / 254)
inches 
# 0.5인치이상 온 날은 몇일입니까?
inches > 0.5
print(np.sum(inches > 0.5))
# 0.5인치이상 온 날의 강수량은?
inches[inches > 0.5]
print(inches[inches > 0.5].sum())
print(np.sum(inches[inches > 0.5]))
# 강수량이 0.5에서 0.8 사이인 날은 몇일이나 되나요?
(inches > 0.5) & (inches < 0.8)
print(np.sum((inches > 0.5) & (inches < 0.8)))
# 강수량이 0.5에서 0.8 사이인 날의 강수량은?
print(np.sum(inches[(inches > 0.5) & (inches < 0.8)]))

#강수량이 0.5보다 작고 0.8이 온날은?
print(np.sum(~((inches > 0.5) & (inches < 0.8))))
(inches < 0.5) | (inches > 0.8)
print(np.sum((inches < 0.5) | (inches > 0.8)))
#강수량이 0.5보다 작고 0.8이 온날의 강수량은?
print(np.sum(inches[(inches < 0.5) | (inches > 0.8)]))
#비가 안온 날
print(np.sum(inches==0))
#비가 온 날
print(np.sum(inches!=0))
```
